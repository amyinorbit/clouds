```python
import numpy as np
from matplotlib.pyplot import *
%matplotlib notebook
```

```python
gamma = 7/5 # Don't worry too much about this -- it's a thermodynamics constant
h_s = 7.99e3 # This is the scale height of the atmosphere
t_s = 273 + 15 # The temperature at sea level (15C) in Kelvin
p_s = 101325 # The pressure at sea level in pascals (Pa)
rho_s = 1.225 # The density at sea level in kg/m^3
rcParams['figure.figsize'] = [10, 3]
```

```python
z = np.linspace(1, 25e3) # This creates an array of values for the altitude, from 0 to 25 km

# For the folowing lines, python allows us to be fancy and automatically calculate
# the values for each value in z. In C++ or C#, you would probably use a loop

# we compute the temperature
t0 = t_s * (1 - ((gamma-1)/gamma) * (z/h_s)) 

# We just compute that once since it's reused for pressure and temperature
common = 1 - ((gamma-1)/gamma) * (z/h_s)

# Then we compute temperatuure and pressure
p0 = p_s * common ** (gamma/(gamma-1))
rho0 = rho_s * common ** (1/(gamma-1))

# and we plot all of this
f, (a1, a2, a3) = subplots(1, 3, sharey=True)
a1.plot(rho0, z)
a1.set_xlabel('density')
a2.plot(t0, z)
a2.set_xlabel('temperature')
a3.plot(p0, z)
a3.set_xlabel('pressure')
f.tight_layout()
show()
```
